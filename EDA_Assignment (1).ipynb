# TASK 2

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['age'], kde=True)
plt.title('Age Distribution')
plt.show()

sns.histplot(df['watch_hours'], kde=True, color = 'yellow')
plt.title('Watch Hours Distribution')
plt.show()

sns.histplot(df['monthly_fee'], kde=True)
plt.title('Monthly Fee Distribution')
plt.show()

sns.countplot(x='churned', data=df)
plt.title('Churn Count')
plt.show()

sns.countplot(x='subscription_type', data=df)
plt.title('Subscription Type Count')
plt.show()

sns.countplot(x='gender', data=df)
plt.title('Gender Count')
plt.show()

sns.countplot(x='region', data=df)
plt.title('Region Count')
plt.xticks(rotation=45)
plt.show()

sns.countplot(x='device', data=df)
plt.title('Device Usage Count')
plt.show()

sns.countplot(x='payment_method', data=df)
plt.title('Payment Method Count')
plt.xticks(rotation=45)
plt.show()

sns.countplot(x='favorite_genre', data=df)
plt.title('Favorite Genre Count')
plt.xticks(rotation=45)
plt.show()
```

```python

sns.barplot(x='subscription_type', y='watch_hours', data=df,color ='green')
plt.title('Avg Watch Hours by Subscription Type')
plt.show()

sns.barplot(x='subscription_type', y='monthly_fee', data=df)
plt.title('Monthly Fee by Subscription Type')
plt.show()

sns.barplot(x='region', y='watch_hours', data=df)
plt.title('Avg Watch Hours by Region')
plt.xticks(rotation=45)
plt.show()

sns.barplot(x='region', y='monthly_fee', data=df)
plt.title('Monthly Fee by Region')
plt.xticks(rotation=45)
plt.show()

sns.barplot(x='device', y='watch_hours', data=df)
plt.title('Avg Watch Hours by Device')
plt.show()

sns.barplot(x='device', y='monthly_fee', data=df)
plt.title('Monthly Fee by Device')
plt.show()

sns.barplot(x='favorite_genre', y='avg_watch_time_per_day', data=df)
plt.title('Avg Watch Time per Day by Genre')
plt.xticks(rotation=45)
plt.show()

sns.barplot(x='gender', y='churned', data=df ,color = 'pink')
plt.title('Churn Rate by Gender')
plt.show()

sns.barplot(x='region', y='churned', data=df)
plt.title('Churn Rate by Region')
plt.xticks(rotation=45)
plt.show()

sns.barplot(x='subscription_type', y='churned', data=df)
plt.title('Churn Rate by Subscription Type')
plt.show()

sns.barplot(x='payment_method', y='churned', data=df)
plt.title('Churn Rate by Payment Method')
plt.xticks(rotation=45)
plt.show()
```

```python
# Task 4: Correlation Analysis

import seaborn as sns
import matplotlib.pyplot as plt

corr = df[['age', 'watch_hours', 'last_login_days', 'monthly_fee',
           'number_of_profiles', 'avg_watch_time_per_day', 'churned']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
